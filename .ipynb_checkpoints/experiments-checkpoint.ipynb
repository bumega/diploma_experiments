```python
import numpy as np
import pandas as pd
from time import perf_counter

from sklearn.datasets import make_classification
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, average_precision_score
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import HistGradientBoostingClassifier

from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression
from umap import UMAP

from pyHSICLasso import HSICLasso

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset, random_split, Subset, Dataset, ConcatDataset


import random
```

```python
def compute_feature_bounds(dataset: Dataset):
    xs = []
    n = len(dataset)
    for i in range(n):
        x, _ = dataset[i]
        xs.append(x)
    X = torch.stack(xs, dim=0)
    return X.min(dim=0).values, X.max(dim=0).values 
```

```python
class UniformNoiseDataset(Dataset):
    def __init__(self, feature_min: torch.Tensor, feature_max: torch.Tensor, n: int, label: int = 0, dtype: torch.dtype | None = None):
        assert feature_min.shape == feature_max.shape
        self.feature_min = feature_min.to(dtype=dtype)
        self.feature_max = feature_max.to(dtype=dtype)
        self.n = int(n)
        self.label = int(label)
        self.dtype = dtype if dtype is not None else torch.float32

    def __len__(self) -> int:
        return self.n

    def __getitem__(self, idx: int):
        u = torch.rand_like(self.feature_min, dtype=self.dtype)
        x = self.feature_min + u * (self.feature_max - self.feature_min)
        y = torch.tensor(self.label, dtype=torch.long)
        return x, y
```

```python
class LabelMappedSubset(Dataset):
    def __init__(self, subset: Subset, label_map: dict[int, int]):

        self.subset = subset
        self.label_map = label_map

    def __getitem__(self, idx):
        x, y = self.subset[idx]
        y_mapped = self.label_map.get(int(y), int(y))
        return x, torch.tensor(y_mapped, dtype=y.dtype)

    def __len__(self):
        return len(self.subset)
```

```python
class Abs(nn.Module):
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return torch.abs(x)
```

```python
class MLP(nn.Module):
    def __init__(self, d_in: int, d_hidden: int = 10, n_hidden_layers: int = 2):

        super().__init__()

        layers = []

        layers.append(nn.Linear(d_in, d_hidden))
        layers.append(Abs())


        for _ in range(n_hidden_layers - 1):
            layers.append(nn.Linear(d_hidden, d_hidden))
            layers.append(Abs())

        layers.append(nn.Linear(d_hidden, 1))

        self.net = nn.Sequential(*layers)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x)
```

```python
def train_one_model(model: nn.Module, dataset_pos_as_1: Dataset, feature_min: torch.Tensor, feature_max: torch.Tensor,
                    num_epochs: int = 20, batch_size: int = 128,
                    lr: float = 3e-3, weight_decay: float = 1e-2, seed: int = 0):


    torch.manual_seed(seed)

    n_real = len(dataset_pos_as_1)
    n_noise = max(1, n_real)
    noise_ds = UniformNoiseDataset(feature_min, feature_max, n_noise, label=0, dtype=torch.float32)


    train_ds = ConcatDataset([dataset_pos_as_1, noise_ds])
    loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, drop_last=False)

    criterion = nn.MSELoss()
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)

    for epoch in range(num_epochs):
        model.train()
        total = 0.0
        for xb, yb in loader:
            xb = xb.float()
            yb = yb.unsqueeze(-1).float()

            pred = model(xb)
            loss = criterion(pred, yb)

            opt.zero_grad(set_to_none=True)
            loss.backward()
            opt.step()

            total += loss.item() * xb.size(0)

        if (epoch + 1) % max(1, num_epochs // 5) == 0:
            avg = total / len(train_ds)
            print(f"[epoch {epoch+1:3d}] loss={avg:.4f}")

    return model
```

```python
def model_train(d_hidden: int, n_hidden_layers: int, train_data: Dataset, num_epochs: int,
                batch_size: int = 128, seed: int = 0,
                output: bool = False):

    torch.manual_seed(seed)

    labels = torch.tensor([int(train_data[i][1]) for i in range(len(train_data))])
    pos_idx = (labels == 1).nonzero(as_tuple=True)[0]
    neg_idx = (labels == -1).nonzero(as_tuple=True)[0]

    train_pos = Subset(train_data, pos_idx)               
    train_neg = Subset(train_data, neg_idx)                 
    train_neg_as_pos = LabelMappedSubset(train_neg, {-1: 1})  


    x0, _ = train_data[0]
    d_in = x0.shape[-1]


    fmin, fmax = compute_feature_bounds(train_data)


    model_pos = MLP(d_in=d_in, d_hidden=d_hidden, n_hidden_layers=n_hidden_layers)
    model_neg = MLP(d_in=d_in, d_hidden=d_hidden, n_hidden_layers=n_hidden_layers)


    model_pos = train_one_model(model_pos, train_pos, fmin, fmax,
                                num_epochs=num_epochs, batch_size=batch_size,
                                seed=seed)
    model_neg = train_one_model(model_neg, train_neg_as_pos, fmin, fmax,
                                num_epochs=num_epochs, batch_size=batch_size,
                                seed=seed + 1)

    if output:
        print("Done. Trained model_pos & model_neg.")

    return model_pos, model_neg
```

```python
def metrics_from_sheet(X, y, model_pos, model_neg, beta_pos=0.1, beta_neg=0.1):
    model_pos.eval(); model_neg.eval()
    with torch.no_grad():
        s1 = model_pos(X).flatten() 
        s2 = model_neg(X).flatten() 

    pos = (y == 1)
    neg = (y == -1)
    n1 = pos.sum().item()
    n2 = neg.sum().item()


    n02_1 = (((s1 >= beta_pos) & (s2 <  beta_neg)) & pos).sum().item()
    n12_1 = (((s1 >= beta_pos) & (s2 >= beta_neg)) & pos).sum().item()


    n02_2 = (((s2 >= beta_neg) & (s1 <  beta_pos)) & neg).sum().item()
    n12_2 = (((s2 >= beta_neg) & (s1 >= beta_pos)) & neg).sum().item()

    a1 = n02_1 / n1 if n1 else 0.0
    a2 = n02_2 / n2 if n2 else 0.0
    F12 = 2 * a1 * a2 / (a1 + a2 + 1e-12)  


    b1 = n12_1 / n1 if n1 else 0.0
    b2 = n12_2 / n2 if n2 else 0.0
    G12 = 2 * b1 * b2 / (b1 + b2 + 1e-12) 

    return dict(n1=n1, n2=n2, n02_1=n02_1, n12_1=n12_1, n02_2=n02_2, n12_2=n12_2,
                a1=a1, a2=a2, F12=F12, b1=b1, b2=b2, G12=G12)
```

```python
def choose_coord(
    X: torch.Tensor,
    y: torch.Tensor,
    d_hidden: int,
    n_hidden_layers: int,
    num_epochs: int,
    batch_size: int = 128,
    seed: int = 0,
    beta_pos: float = 0.1,
    beta_neg: float = 0.1,
    output: bool = False,
):

    n_features = X.shape[1]

    rows = []
    best_idx, best_key = None, -float("inf")

    for i in range(n_features):
        X_upd = torch.cat((X[:, :i], X[:, i+1:]), dim=1)

        train_data_upd = TensorDataset(X_upd, y)
        model_pos, model_neg = model_train(
            train_data=train_data_upd,
            d_hidden=d_hidden,
            n_hidden_layers=n_hidden_layers,
            num_epochs=num_epochs,
            batch_size=batch_size,
            seed=seed,
            output=output,
        )

        m = metrics_from_sheet(
            X_upd, y, model_pos, model_neg, beta_pos=beta_pos, beta_neg=beta_neg
        )
        m["removed_idx"] = i
        m["S"] = m["F12"] - 0.5 * m["G12"]   
        rows.append(m)

        key = m["S"]
        if key > best_key:
            best_key = key
            best_idx = i

    df = pd.DataFrame(rows).sort_values(["S", "F12", "G12"], ascending=[False, False, True]).reset_index(drop=True)
    return best_idx, df
```

```python
def set_seed(seed=42):
    random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed=seed)
set_seed()
```

```python
def make_dataset_A(seed=42):
    X, y = make_classification(
        n_samples=3000, n_features=8, n_informative=3, n_redundant=2,
        class_sep=1.2, flip_y=0.02, random_state=seed)
    X[:, 0] = np.sin(X[:, 0]) + 0.1 * X[:, 1]
    X[:, 2] = X[:, 2] ** 2
    y = y.astype(int)
    y = 2*y - 1
    return X.astype(np.float64), y.astype(int)
```

```python
def make_dataset_B(seed=42):
    X, y = make_classification(
        n_samples=6000, n_features=30, n_informative=5, n_redundant=5,
        class_sep=0.8, flip_y=0.01, weights=[0.75, 0.25], random_state=seed)
    X[:, 10] = np.sin(X[:, 0]) * X[:, 1]
    X[:, 11] = X[:, 2] ** 2
    X[:, 12] = np.exp(-X[:, 3]**2)
    X[:, 20:25] = X[:, :5] + 0.05 * np.random.randn(len(X), 5)
    y = y.astype(int)
    y = 2*y - 1
    return X.astype(np.float64), y.astype(int)
```

```python
class ReducerBase:
    def fit(self, X, y=None): ...
    def transform(self, X): ...
    def fit_transform(self, X, y=None):
        self.fit(X, y); return self.transform(X)
```

```python
class PCAReducer(ReducerBase):
    def __init__(self, n_components): self.pca = PCA(n_components=n_components, random_state=42)
    def fit(self, X, y=None): self.pca.fit(X); return self
    def transform(self, X): return self.pca.transform(X)
```

```python
class PLSReducer(ReducerBase):

    def __init__(self, n_components): self.pls = PLSRegression(n_components=n_components, scale=False)
    def fit(self, X, y): self.pls.fit(X, y.astype(float)); return self
    def transform(self, X): return self.pls.transform(X)
```

```python
class UMAPReducer(ReducerBase):
    def __init__(self, n_components):
        self.umap = UMAP(n_components=n_components, random_state=42, n_neighbors=15, min_dist=0.1, target_metric='categorical')
    def fit(self, X, y): self.umap.fit(X, y); return self
    def transform(self, X): return self.umap.transform(X)
```

```python
class HSICSelector(ReducerBase):
    def __init__(self, n_select):
        self.n_select = n_select
        self.model = HSICLasso()
        self.mask_ = None

    def fit(self, X, y):

        y = np.asarray(y).ravel().astype(int)
        self.model.input(X, y)

        self.model.classification(self.n_select)

        idx = np.array(self.model.get_index(), dtype=int)
        self.mask_ = np.zeros(X.shape[1], dtype=bool)
        self.mask_[idx] = True
        return self

    def transform(self, X):
        return X[:, self.mask_]
```

```python
class MLPReducer(ReducerBase):
    def __init__(self, d_hidden, n_hidden_layers, num_epochs, batch_size, n_select, seed=0, output=False):
        self.d_hidden = d_hidden
        self.n_hidden_layers = n_hidden_layers
        self.num_epochs = num_epochs
        self.batch_size = batch_size
        self.seed = seed
        self.output = output
        self.n_select = n_select

        self.selected_indices_ = None
        self.removed_indices_ = None
        self.history_ = []  

    def _to_torch2d(self, X):
        """Приводим к torch.float32 2D tensor без копии если возможно."""
        if isinstance(X, torch.Tensor):
            assert X.ndim == 2, "X must be 2D"
            return X.float()
        X = np.asarray(X)
        assert X.ndim == 2, "X must be 2D"
        return torch.from_numpy(X).float()

    def _to_torch1d_int(self, y):
        if isinstance(y, torch.Tensor):
            assert y.ndim == 1, "y must be 1D"
            return y.to(dtype=torch.int64)
        y = np.asarray(y).ravel()
        return torch.from_numpy(y.astype(np.int64))

    def fit(self, X, y):

        X_t = self._to_torch2d(X)
        y_t = self._to_torch1d_int(y)

        d = X_t.shape[1]
        keep = list(range(d))
        removed = []


        while len(keep) > self.n_select:

            X_cur = X_t[:, keep]

            best_idx, df = choose_coord(
                X=X_cur, y=y_t,
                d_hidden=self.d_hidden,
                n_hidden_layers=self.n_hidden_layers,
                num_epochs=self.num_epochs,
                batch_size=self.batch_size,
                seed=self.seed,
                output=self.output,
            )

            removed_global = keep[best_idx]
            removed.append(removed_global)
            del keep[best_idx]


            self.history_.append({
                "removed_global": removed_global,
                "n_left": len(keep),
                "df": df, 
            })

        self.selected_indices_ = keep
        self.removed_indices_ = removed
        return self

    def transform(self, X):
        assert self.selected_indices_ is not None, "Call fit() before transform()."

        if isinstance(X, torch.Tensor):
            return X[:, self.selected_indices_]
        X = np.asarray(X)
        return X[:, self.selected_indices_]


    def fit_transform(self, X, y):
        return self.fit(X, y).transform(X)


    def get_support(self, indices=False):
        assert self.selected_indices_ is not None, "Call fit() first."
        if indices:
            return np.array(self.selected_indices_, dtype=int)
        mask = np.zeros(max(max(self.selected_indices_)+1, 0) if self.selected_indices_ else 0, dtype=bool)
        if len(mask) == 0:
            mask = np.zeros(self.n_select, dtype=bool) 
        mask[self.selected_indices_] = True
        return mask
```

```python
class EnsembleReducer(ReducerBase):
    def __init__(self, d_hidden, n_hidden_layers, num_epochs, batch_size, n_select, num_models, num_attempts, 
                 num_coords, num_samples, seed=0, output=False):
        self.d_hidden = d_hidden
        self.n_hidden_layers = n_hidden_layers
        self.num_epochs = num_epochs
        self.batch_size = batch_size
        self.seed = seed
        self.output = output
        self.n_select = n_select
        self.num_samples = num_samples
        self.num_models = num_models
        self.num_attempts = num_attempts
        self.num_coords = num_coords

        self.list_model_pos = []
        self.list_model_neg = []
        self.list_model_subs = []
        self.coord_prob = None
        self.history_ = []  
        self.rng = np.random.default_rng()

    def _to_torch2d(self, X):

        if isinstance(X, torch.Tensor):
            assert X.ndim == 2, "X must be 2D"
            return X.float()
        X = np.asarray(X)
        assert X.ndim == 2, "X must be 2D"
        return torch.from_numpy(X).float()

    def _to_torch1d_int(self, y):
        if isinstance(y, torch.Tensor):
            assert y.ndim == 1, "y must be 1D"
            return y.to(dtype=torch.int64)
        y = np.asarray(y).ravel()
        return torch.from_numpy(y.astype(np.int64))
    
    def fit(self, X, y):

        X_t = self._to_torch2d(X)
        y_t = self._to_torch1d_int(y)
        n = X_t.shape[1]
        self.coord_prob = np.array(([1 / n for i in range(n)]), dtype=float)

        for k in range(self.num_samples):
            for i in range(self.num_models):
                best_model_pos = None
                best_model_neg = None
                best_subs_idx = None
                best_loss = float('inf')
                for j in range(self.num_attempts):
                    subs_idx = np.sort(self.rng.choice(self.coord_prob.size, size=self.num_coords, replace=False, p=self.coord_prob))
                    X_sub = X_t[:, subs_idx]
                    train_ds = torch.utils.data.TensorDataset(X_sub, y_t)
                    model_pos, model_neg = model_train(d_hidden=self.d_hidden, n_hidden_layers=self.n_hidden_layers, 
                                                       train_data=train_ds, num_epochs=self.num_epochs,
                                                       batch_size=self.batch_size)
                    model_pos.eval()
                    model_neg.eval()
                    
                    with torch.no_grad():
                        output_pos = model_pos(X_sub).squeeze(-1)
                        output_neg = model_neg(X_sub).squeeze(-1)
                    mask_pos = (output_pos >= 0.1) & (output_neg < 0.1)
                    mask_neg = (output_neg >= 0.1) & (output_pos < 0.1)
                    mask_undef = ((output_neg >= 0.1) & (output_pos >= 0.1)) | ((output_neg < 0.1) & (output_pos < 0.1))
                    loss = ((torch.sum((y_t[mask_pos] - 1) ** 2) + torch.sum((y_t[mask_neg] + 1) ** 2) + torch.sum((y_t[mask_undef]) ** 2)) / y_t.numel()).item()
                    if loss < best_loss:
                        best_loss = loss
                        best_subs_idx = subs_idx
                        best_model_pos = model_pos
                        best_model_neg = model_neg
                self.list_model_pos.append(best_model_pos)
                self.list_model_neg.append(best_model_neg)
                self.list_model_subs.append(best_subs_idx)
            coord_new_prob = np.zeros(n)
            for i in range(len(self.list_model_pos) - self.num_models, len(self.list_model_pos)):
                for ind in self.list_model_subs[i]:
                    coord_new_prob[ind] += 1
            coord_new_prob /= self.num_models
            eta = coord_new_prob.astype(float)                    
            p = eta.size
            C0 = 1.0                                               
            thr = C0 / np.log(max(p, 3))             


            eta_tilde = np.where(eta > thr, eta, C0 / p)


            self.coord_prob = eta_tilde / eta_tilde.sum()
    
    def coord_prob_get(self):
        return self.coord_prob

    def transform(self, X):
        assert self.coord_prob is not None, "Call fit() before transform()."
        assert hasattr(self, "n_select"), "n_select must be set in __init__()."


        k = min(self.n_select, len(self.coord_prob))
        topk_idx = np.argsort(self.coord_prob)[-k:] 
        topk_idx = np.sort(topk_idx)              
        self.selected_indices_ = topk_idx


        if isinstance(X, torch.Tensor):
            assert X.ndim == 2, "X must be 2D tensor"
            return X[:, topk_idx]
        else:
            X = np.asarray(X)
            assert X.ndim == 2, "X must be 2D array"
            return X[:, topk_idx]
        

```

```python
def get_models():
    return {
        "MLP": MLPClassifier(hidden_layer_sizes=(64,32), activation='relu',
                             alpha=1e-4, batch_size=256, learning_rate_init=1e-3,
                             max_iter=200, early_stopping=True, n_iter_no_change=10, random_state=42),
        "HGBT": HistGradientBoostingClassifier(random_state=42)
    }
```

```python
def evaluate_reduction(X, y, reducer_name, reducer_ctor, q, dataset_name,
                       n_splits=5, reducer_kwargs=None):

    reducer_kwargs = reducer_kwargs or {}
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    rows = []

    for fold, (tr, te) in enumerate(skf.split(X, y), 1):
        Xtr, Xte = X[tr], X[te]
        ytr, yte = y[tr], y[te]

        scaler = StandardScaler()
        Xtr_s = scaler.fit_transform(Xtr)
        Xte_s = scaler.transform(Xte)

        # БАЗА: без редукции
        for mname, base_model in get_models().items():
            m0 = base_model.__class__(**base_model.get_params())
            t0 = perf_counter(); m0.fit(Xtr_s, ytr); t0 = perf_counter() - t0
            p0 = m0.predict_proba(Xte_s)[:,1] if hasattr(m0, "predict_proba") else m0.decision_function(Xte_s)
            auc0 = roc_auc_score(yte, p0); ap0 = average_precision_score(yte, p0)

            # С РЕДУКЦИЕЙ
            red = reducer_ctor(q=q, **reducer_kwargs)  # ← передаём доп. параметры сюда
            tR = perf_counter()
            Ztr = red.fit_transform(Xtr_s, ytr)
            Zte = red.transform(Xte_s)
            tR = perf_counter() - tR

            m1 = base_model.__class__(**base_model.get_params())
            t1 = perf_counter(); m1.fit(Ztr, ytr); t1 = perf_counter() - t1
            p1 = m1.predict_proba(Zte)[:,1] if hasattr(m1, "predict_proba") else m1.decision_function(Zte)
            auc1 = roc_auc_score(yte, p1); ap1 = average_precision_score(yte, p1)

            rows.append({
                "dataset": dataset_name, "method": reducer_name, "q": q, "model": mname, "fold": fold,
                "AUC": auc1, "PR_AUC": ap1, "ΔAUC": auc1-auc0, "ΔPR_AUC": ap1-ap0,
                "t_reducer": tR, "t_model": t1, "t_model_base": t0
            })
    return rows
```

```python
def run_all_B():
    configs = [
        #("DS-A", make_dataset_A, [2,3,5]),
        ("DS-B", make_dataset_B, [5, 10, 15]),
    ]
    methods = [
        ("PCA", lambda q: PCAReducer(q)),
        ("PLS", lambda q: PLSReducer(q)),
        ("UMAP_sup", lambda q: UMAPReducer(q)),
        ("HSIC_Lasso", lambda q: HSICSelector(q)),
        #("MLP_unar", lambda q: MLPReducer(
        #    d_hidden=64,
        #    n_hidden_layers=2,
        #    num_epochs=40,
        #    batch_size=256,
        #    n_select=q,
        #    seed=0,
        #    output=False,
        #)),
        ("Ensembleunar", lambda q: EnsembleReducer(
            d_hidden=32, n_hidden_layers=1,
            num_epochs=5, batch_size=1024,
            n_select=q,
            num_models=8,     # B1
            num_attempts=7,   # B2
            num_coords=4,      # d
            num_samples=4,     # итераций RaSE
            seed=42, output=True
        )),
    ]
    all_rows = []
    for dname, maker, qgrid in configs:
        X, y = maker()
        for name, ctor in methods:
            for q in qgrid:
                all_rows += evaluate_reduction(X, y, name, ctor, q, dname)
    df = pd.DataFrame(all_rows)
    # агрегаты по фолдам
    agg = (df.groupby(["dataset","method","q","model"])
             .agg(AUC_mean=("AUC","mean"),
                  PR_AUC_mean=("PR_AUC","mean"),
                  dAUC_mean=("ΔAUC","mean"),
                  dPR_mean=("ΔPR_AUC","mean"),
                  t_red_med=("t_reducer","median"),
                  t_mod_med=("t_model","median"))
             .reset_index()
             .sort_values(["dataset","model","dAUC_mean"], ascending=[True, True, False]))
    return df, agg
```
